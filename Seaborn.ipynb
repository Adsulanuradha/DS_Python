```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
#seaborn plots documentation 
#strip plot
sns.stripplot(y='tip',data=tips,jitter=True)  
#plt.show()
```

```python
#grouping with stripplot()
sns.stripplot(x="sex",y="tip",data=tips,jitter=True)
```

```python
sns.stripplot(x="day",y="tip",data=tips,jitter=True)
```

```python
sns.stripplot(x="day",y="tip",data=tips,size=2,jitter=True)
```

```python
#swarm plot
sns.swarmplot(x='day',y='tip',data=tips,size=6)
```

```python
#high amount of tip
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='time')
```

```python
#subplots
plt.subplot(2,2,1)
sns.boxplot(x='day',y='tip',data=tips)

plt.subplot(2,2,2)
sns.boxplot(x='day',y='tip',data=tips)
plt.title=("Boxplot of tips")

plt.subplot(2,2,3)
sns.violinplot(x='day',y='tip',data=tips)

plt.subplot(2,2,4)
sns.violinplot(x='day',y='tip',data=tips)
plt.title=("violin plot of tips ")

plt.show()
```

```python
#combiningg plots
sns.violinplot(x='day',y='tip',data=tips,inner='stick',color='purple')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time')
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex',orient='h')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2) #empty spaces for plots ,change nrows and ncols
```

```python
#joint plot -scatter plot and individual histogram of x and y variables
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
pd.plotting.scatter_matrix(tips)
```

```python
#pair plot: if dataset has many numeric column you can plot pairplot
sns.pairplot(tips)
```

```python
df.loc[df['Survived']==1,'Age']  #age col contains some na values
```

```python
survived=df.loc[df['Survived']==1,'Age'].dropna()
sns.distplot(survived)
```

```python
not_survived=df.loc[df['Survived']==0,'Age'].dropna()
sns.distplot(not_survived)
```

```python
#pclass vs survive
grid=sns.FacetGrid(df,col="Survived",row="Pclass",height=2.4,aspect=1.5)
grid.map(plt.hist,'Age',alpha=0.5,bins=20)
grid.add_legend() 
```

```python
figbi,axesbi=plt.subplots(2,4,figsize=(18,10))
df.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
df.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
df.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
df.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
df.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x='Survived',y='Age',data=df,ax=axesbi[1,1])
sns.boxplot(x='Survived',y='Fare',data=df,ax=axesbi[1,2])
plt.show()
figbi.savefig("my1stfig.png")
from google.colab import files
files.download("my1stfig.png")
```
